```python
#Face detection using sklearn
from sklearn.model_selection import GridSearchCV
import numpy as np
from sklearn.datasets import fetch_lfw_people
lfw_people = fetch_lfw_people(min_faces_per_person=70, resize=0.4)
print ("Data importing done")
```

```python
import matplotlib.pyplot as plt
print "No. of images " +str(lfw_people.images.shape[0])+" with dimensions "+str(lfw_people.images.shape[1:3])
n_samples, h, w = lfw_people.images.shape
X= lfw_people.data
n_features=X.shape[1]
print "No of features are "+str(n_features)
y = lfw_people.target
target_names = lfw_people.target_names
print "No of classes are " + str(target_names.shape)
```

```python
#Data splitting
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42)
print "Training data has shape "+ str(X_train.shape)

plot_gallery(X_train,h, w)
print "Certain instances of training set are shown below"
def plot_gallery(images,h, w, n_row=3, n_col=4):
    plt.figure(figsize=(1.8 * n_col, 2.4 * n_row))
    plt.subplots_adjust(bottom=0, left=.01, right=.99, top=.90, hspace=.35)
    for i in range(n_row * n_col):
        plt.subplot(n_row, n_col, i + 1)
        plt.imshow(images[i].reshape((h, w)), cmap=plt.cm.gray)
        #plt.title(titles[i], size=12)
        plt.xticks(())
        plt.yticks(())
def title(y_pred, y_test, target_names, i):
    pred_name = target_names[y_pred[i]].rsplit(' ', 1)[-1]
    true_name = target_names[y_test[i]].rsplit(' ', 1)[-1]
    return 'predicted: %s\ntrue:      %s' % (pred_name, true_name)
plt.show()

```

```python
#Computing PCA
from sklearn.decomposition import PCA
from sklearn import svm
n_components = 150
print "Extracting " +str(n_components)+" eigenfaces from "+ str(X_train.shape[0])+" faces"
#The 1st eigen face has the most variation, 2 nd has the secondmost variation an so on.
pca = PCA(n_components=n_components, svd_solver='randomized',
          whiten=True).fit(X_train)
print "Originally pca or eigenfaces components got converted into " +str(pca.components_.shape)
eigenfaces = pca.components_.reshape((n_components, h, w))
print "After reshaping pca or eigenfaces components got converted into " +str(eigenfaces.shape)
#Transformation data into pca
X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)
plot_gallery(eigenfaces,h, w, n_row=3, n_col=4)
print "Done"
```

```python
#Training a classifier now
from sklearn.model_selection import GridSearchCV
parameters = {'kernel':('linear', 'rbf'), 'C':[1, 10]}
svc = svm.SVC()
clf = GridSearchCV(svc, parameters)
clf.fit(X_train_pca, y_train)
y_pred = clf.predict(X_test_pca)
from sklearn.metrics import accuracy_score
print "Accuracy score is "+ str(accuracy_score(y_test,y_pred))
```
